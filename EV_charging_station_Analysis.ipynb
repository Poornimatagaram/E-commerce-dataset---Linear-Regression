```python
import pandas as pd

# Google Drive CSV public file link
file_id = '1LRFFnemGnTRkq8-K2ZnkXp_w6RLREgk-'
url = f'https://drive.google.com/uc?id={file_id}'

# Load the CSV
df = pd.read_csv(url)

# Quick dataset overview
print("Dataset Overview:")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")

# Display first 10 rows
print("\n First 10 rows:")
df.head(10)
```

```python
# Step 3: Filter for Electric Vehicle Stations Only

# Check unique fuel types in the dataset
print(" Fuel Types in Dataset:")
print(df['Fuel Type Code'].value_counts())

# Filter for Electric Vehicle stations only
# EV stations should have 'ELEC' as Fuel Type Code
ev_df = df[df['Fuel Type Code'] == 'ELEC'].copy()

print(f"\n After filtering for EV stations:")
print(f"Original dataset: {df.shape[0]} records")
print(f"EV stations only: {ev_df.shape[0]} records")
print(f"Percentage of EV stations: {(ev_df.shape[0]/df.shape[0]*100):.1f}%")

# Quick check of EV data
print(f"\n EV Dataset Overview:")
print(f"Shape: {ev_df.shape}")
ev_df.head(3)
```

```python
# Step 4: Examine Key EV Columns for Analysis

# Key columns for our analysis
key_cols = ['State', 'City', 'Latitude', 'Longitude', 'Status Code',
           'EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count',
           'EV Connector Types', 'Access Code', 'Open Date', 'EV Network']

print(" Key EV Columns Overview:")
for col in key_cols:
    print(f"\n{col}:")
    print(f"  Non-null values: {ev_df[col].notna().sum()} / {len(ev_df)}")
    if col in ['EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count']:
        print(f"  Mean: {ev_df[col].mean():.2f}")
        print(f"  Max: {ev_df[col].max()}")
    elif col in ['State', 'Status Code', 'Access Code']:
        print(f"  Unique values: {ev_df[col].value_counts().head(3).to_dict()}")

# Check for missing coordinates (crucial for mapping)
print(f"\n Geographic Data Quality:")
print(f"Complete lat/lng pairs: {ev_df[['Latitude', 'Longitude']].notna().all(axis=1).sum()}")
print(f"Missing coordinates: {ev_df[['Latitude', 'Longitude']].isna().any(axis=1).sum()}")
```

```python
# Step 5: Data Cleaning - Remove Unnecessary Columns & Handle Nulls

# First, let's identify which columns we actually need for EV analysis
essential_columns = [
    # Geographic info
    'State', 'City', 'ZIP', 'Latitude', 'Longitude',

    # Station details
    'Station Name', 'Street Address', 'Status Code', 'Access Code',

    # EV-specific charging info
    'EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count',
    'EV Connector Types', 'EV Network', 'EV Pricing',

    # Temporal info
    'Open Date', 'Date Last Confirmed',

    # Access and operational info
    'Groups With Access Code', 'Access Days Time', 'Owner Type Code',
    'Facility Type'
]

# Create clean dataset with only essential columns
ev_clean = ev_df[essential_columns].copy()

print(" Data Cleaning Results:")
print(f"Reduced from {ev_df.shape[1]} to {ev_clean.shape[1]} columns")
print(f"Kept {len(ev_clean)} EV stations")

# Check missing values in essential columns
print(f"\n Missing Values in Essential Columns:")
missing_summary = ev_clean.isnull().sum()
missing_pct = (missing_summary / len(ev_clean) * 100).round(1)

for col in essential_columns:
    if missing_summary[col] > 0:
        print(f"{col}: {missing_summary[col]} missing ({missing_pct[col]}%)")
    else:
        print(f"{col}:  Complete")

# Show the cleaned dataset
print(f"\n Cleaned Dataset Shape: {ev_clean.shape}")
ev_clean.head(3)
```

```python
# Step 6: Smart Null Handling for EV Analysis

# For EV charging data, null often means "0 of that charging type"
print(" Smart Null Handling:")

# Fill charging counts: null = 0 (station doesn't have that charging type)
ev_clean['EV Level1 EVSE Num'] = ev_clean['EV Level1 EVSE Num'].fillna(0)
ev_clean['EV Level2 EVSE Num'] = ev_clean['EV Level2 EVSE Num'].fillna(0)
ev_clean['EV DC Fast Count'] = ev_clean['EV DC Fast Count'].fillna(0)

# Fill other important nulls
ev_clean['EV Network'] = ev_clean['EV Network'].fillna('Non-Networked')
ev_clean['EV Pricing'] = ev_clean['EV Pricing'].fillna('Unknown')
ev_clean['Access Days Time'] = ev_clean['Access Days Time'].fillna('Hours Unknown')

# Create total charging ports column
ev_clean['Total_Charging_Ports'] = (ev_clean['EV Level1 EVSE Num'] +
                                   ev_clean['EV Level2 EVSE Num'] +
                                   ev_clean['EV DC Fast Count'])

# Create charging type categories
def get_charging_type(row):
    if row['EV DC Fast Count'] > 0:
        return 'DC Fast Charging'
    elif row['EV Level2 EVSE Num'] > 0:
        return 'Level 2 Charging'
    elif row['EV Level1 EVSE Num'] > 0:
        return 'Level 1 Charging'
    else:
        return 'Unknown'

ev_clean['Primary_Charging_Type'] = ev_clean.apply(get_charging_type, axis=1)

# Handle other nulls
ev_clean['EV Network'] = ev_clean['EV Network'].fillna('Non-Networked')
ev_clean['EV Pricing'] = ev_clean['EV Pricing'].fillna('Unknown')

# Final data quality check
print(" After cleaning:")
print(f"Dataset shape: {ev_clean.shape}")
print(f"Null values remaining: {ev_clean.isnull().sum().sum()}")

print(f"\n Charging Type Distribution:")
print(ev_clean['Primary_Charging_Type'].value_counts())

print(f"\n Charging Statistics:")
print(f"Average ports per station: {ev_clean['Total_Charging_Ports'].mean():.1f}")
print(f"Stations with DC Fast Charging: {(ev_clean['EV DC Fast Count'] > 0).sum():,}")
print(f"Stations with Level 2: {(ev_clean['EV Level2 EVSE Num'] > 0).sum():,}")

ev_clean.head(3)
```

```python
# Step 7: Geographic Distribution Analysis - INSIGHT 1

print(" INSIGHT 1: Geographic Distribution of EV Charging Stations")
print("="*60)

# Top 10 states by EV station count
top_states = ev_clean['State'].value_counts().head(10)
print(f"\n Top 10 States by EV Station Count:")
for state, count in top_states.items():
    pct = (count/len(ev_clean)*100)
    print(f"{state}: {count:,} stations ({pct:.1f}%)")

# Calculate stations per capita (you'd need population data for exact numbers)
# For now, let's analyze distribution patterns
print(f"\n Geographic Distribution Patterns:")
print(f"Total states/provinces: {ev_clean['State'].nunique()}")
print(f"California dominance: {(ev_clean['State']=='CA').sum()/len(ev_clean)*100:.1f}% of all US EV stations")

# Access patterns by state (top 5 states)
print(f"\n Public vs Private Access (Top 5 States):")
top_5_states = top_states.head(5).index
for state in top_5_states:
    state_data = ev_clean[ev_clean['State'] == state]
    public_pct = (state_data['Access Code'] == 'public').mean() * 100
    print(f"{state}: {public_pct:.1f}% public access")

# Charging type distribution by top states
print(f"\n Charging Types by Top 3 States:")
for state in top_5_states[:3]:
    state_data = ev_clean[ev_clean['State'] == state]
    dc_fast_pct = (state_data['Primary_Charging_Type'] == 'DC Fast Charging').mean() * 100
    level2_pct = (state_data['Primary_Charging_Type'] == 'Level 2 Charging').mean() * 100
    print(f"{state}: {dc_fast_pct:.1f}% DC Fast, {level2_pct:.1f}% Level 2")

# Create summary for dashboard
geographic_summary = {
    'total_stations': len(ev_clean),
    'top_state': top_states.index[0],
    'top_state_count': top_states.iloc[0],
    'states_covered': ev_clean['State'].nunique(),
    'ca_dominance_pct': (ev_clean['State']=='CA').sum()/len(ev_clean)*100
}

print(f"\n KEY INSIGHT 1 SUMMARY:")
print(f"• {geographic_summary['total_stations']:,} EV charging stations analyzed")
print(f"• {geographic_summary['states_covered']} states/provinces covered")
print(f"• {geographic_summary['top_state']} leads with {geographic_summary['top_state_count']:,} stations")
print(f"• California has {geographic_summary['ca_dominance_pct']:.1f}% of all stations - showing infrastructure gaps in other regions")
```

```python
# Step 8: Infrastructure Growth & Trends Analysis - INSIGHT 2

print(" INSIGHT 2: EV Infrastructure Growth Trends & Patterns")
print("="*60)

# Convert Open Date to datetime for analysis
ev_clean['Open Date'] = pd.to_datetime(ev_clean['Open Date'], errors='coerce')
ev_clean['Open Year'] = ev_clean['Open Date'].dt.year

# Remove unrealistic years (before 1990 or after 2024)
valid_years = ev_clean[(ev_clean['Open Year'] >= 1990) & (ev_clean['Open Year'] <= 2024)]

print(f" Stations with valid opening dates: {len(valid_years):,} out of {len(ev_clean):,}")

# Growth by year
yearly_growth = valid_years['Open Year'].value_counts().sort_index()
print(f"\n EV Station Growth by Year (Last 10 Years):")
for year in range(2015, 2025):
    if year in yearly_growth.index:
        count = yearly_growth[year]
        print(f"{year}: {count:,} new stations")

# Growth acceleration analysis
recent_years = yearly_growth[yearly_growth.index >= 2015]
if len(recent_years) > 1:
    growth_rate = ((recent_years.iloc[-1] - recent_years.iloc[0]) / recent_years.iloc[0]) * 100
else:
    growth_rate = 0

print(f"\n Growth Analysis:")
print(f"Peak year: {yearly_growth.idxmax()} with {yearly_growth.max():,} new stations")
print(f"Recent growth (2015-2024): {growth_rate:.1f}% change")
print(f"Major acceleration in 2021: {yearly_growth[2021]:,} stations (COVID recovery + policy push)")

# Network evolution analysis
print(f"\n Network Infrastructure Patterns:")
network_counts = ev_clean['EV Network'].value_counts().head(8)
for network, count in network_counts.items():
    pct = (count/len(ev_clean)*100)
    print(f"{network}: {count:,} stations ({pct:.1f}%)")

# Charging technology trends
print(f"\n Charging Technology Distribution:")
charging_stats = ev_clean['Primary_Charging_Type'].value_counts()
for charge_type, count in charging_stats.items():
    pct = (count/len(ev_clean)*100)
    print(f"{charge_type}: {count:,} stations ({pct:.1f}%)")

# Infrastructure density analysis
print(f"\n Infrastructure Density Insights:")
print(f"Average charging ports per station: {ev_clean['Total_Charging_Ports'].mean():.1f}")
print(f"Stations with 10+ ports: {(ev_clean['Total_Charging_Ports'] >= 10).sum():,}")
print(f"Large stations (20+ ports): {(ev_clean['Total_Charging_Ports'] >= 20).sum():,}")

print(f"\n KEY INSIGHT 2 SUMMARY:")
print(f"• Peak installation year: {yearly_growth.idxmax()} ({yearly_growth.max():,} stations)")
print(f"• Non-Networked stations dominate: {network_counts.iloc[0]:,} stations")
print(f"• Level 2 charging is standard: {charging_stats.iloc[0]:,} stations")
print(f"• Infrastructure is scaling up: {(ev_clean['Total_Charging_Ports'] >= 10).sum():,} multi-port stations")
```

```python
# Step 9: Accessibility & Infrastructure Gaps Analysis - INSIGHT 3

print(" INSIGHT 3: EV Charging Accessibility & Infrastructure Gaps")
print("="*65)

# Access patterns analysis
print(" Public vs Private Access Distribution:")
access_dist = ev_clean['Access Code'].value_counts()
for access, count in access_dist.items():
    pct = (count/len(ev_clean)*100)
    print(f"{access.title()}: {count:,} stations ({pct:.1f}%)")

# Operational status analysis
print(f"\n Operational Status Analysis:")
status_dist = ev_clean['Status Code'].value_counts()
status_labels = {'E': 'Operational', 'T': 'Temporarily Unavailable', 'P': 'Planned'}
for status, count in status_dist.items():
    label = status_labels.get(status, status)
    pct = (count/len(ev_clean)*100)
    print(f"{label}: {count:,} stations ({pct:.1f}%)")

# 24/7 availability analysis
has_hours = ev_clean['Access Days Time'].notna()
always_open = ev_clean['Access Days Time'].str.contains('24 hours', case=False, na=False)
print(f"\n Operating Hours Analysis:")
print(f"Stations with known hours: {has_hours.sum():,} ({(has_hours.mean()*100):.1f}%)")
print(f"24/7 accessible stations: {always_open.sum():,} ({(always_open.mean()*100):.1f}%)")

# Geographic accessibility gaps
print(f"\n Geographic Accessibility Analysis:")

# Bottom 10 states by station count (identifying gaps)
bottom_states = ev_clean['State'].value_counts().tail(10)
print(f"\nStates with LOWEST EV infrastructure (potential gaps):")
for state, count in bottom_states.items():
    print(f"{state}: {count} stations")

# DC Fast Charging gaps by state (top 10 states)
top_states = ev_clean['State'].value_counts().head(10).index
print(f"\n DC Fast Charging Coverage (Top 10 States):")
for state in top_states:
    state_data = ev_clean[ev_clean['State'] == state]
    total_stations = len(state_data)
    dc_stations = (state_data['EV DC Fast Count'] > 0).sum()
    dc_coverage = (dc_stations/total_stations*100)
    print(f"{state}: {dc_coverage:.1f}% of stations have DC Fast Charging ({dc_stations:,}/{total_stations:,})")

# Network coverage gaps
print(f"\n Network Coverage Analysis:")
networked = ev_clean['EV Network'] != 'Non-Networked'
print(f"Networked stations: {networked.sum():,} ({(networked.mean()*100):.1f}%)")
print(f"Non-networked stations: {(~networked).sum():,} ({((~networked).mean()*100):.1f}%)")

# Infrastructure adequacy assessment
print(f"\n KEY INSIGHT 3 SUMMARY:")
print(f"• Public accessibility: {access_dist['public']:,} stations ({(access_dist['public']/len(ev_clean)*100):.1f}%) are publicly accessible")
print(f"• Operational reliability: {status_dist['E']:,} stations ({(status_dist['E']/len(ev_clean)*100):.1f}%) are fully operational")
print(f"• 24/7 availability limited: Only {(always_open.mean()*100):.1f}% offer round-the-clock access")
print(f"• DC Fast Charging gaps: Highway travel still challenging in many regions")
print(f"• Geographic inequality: {bottom_states.iloc[0]} has only {bottom_states.iloc[-1]} stations vs CA's {ev_clean['State'].value_counts().iloc[0]:,}")

# Create accessibility score
public_score = (access_dist['public']/len(ev_clean)*100)
operational_score = (status_dist['E']/len(ev_clean)*100)
networked_score = (networked.mean()*100)
overall_accessibility = (public_score + operational_score + networked_score) / 3

print(f"\n OVERALL ACCESSIBILITY SCORE: {overall_accessibility:.1f}/100")
print("   (Based on public access + operational status + network coverage)")
```

```python
# Step 10: Prepare Clean Data for Tableau Dashboard

print(" Preparing Data for Tableau Dashboard")
print("="*45)

# Create final dataset for Tableau with key columns
tableau_data = ev_clean[[
    # Geographic
    'State', 'City', 'Latitude', 'Longitude',
    # Station Info
    'Station Name', 'Status Code', 'Access Code',
    # Charging Details
    'EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count',
    'Total_Charging_Ports', 'Primary_Charging_Type',
    # Network & Pricing
    'EV Network', 'EV Pricing',
    # Temporal
    'Open Date', 'Open Year',
    # Access
    'Access Days Time'
]].copy()

# Clean up for Tableau
tableau_data['Status'] = tableau_data['Status Code'].map({
    'E': 'Operational',
    'T': 'Temporarily Unavailable',
    'P': 'Planned'
})

tableau_data['Access Type'] = tableau_data['Access Code'].map({
    'public': 'Public',
    'private': 'Private'
})

# Create summary statistics for dashboard
state_summary = ev_clean.groupby('State').agg({
    'Station Name': 'count',
    'EV Level2 EVSE Num': 'sum',
    'EV DC Fast Count': 'sum',
    'Total_Charging_Ports': 'sum'
}).rename(columns={
    'Station Name': 'Total_Stations',
    'EV Level2 EVSE Num': 'Level2_Ports',
    'EV DC Fast Count': 'DC_Fast_Ports',
    'Total_Charging_Ports': 'All_Ports'
}).reset_index()

# Add percentage columns
state_summary['Public_Access_Pct'] = ev_clean.groupby('State')['Access Code'].apply(
    lambda x: (x == 'public').mean() * 100
).values

state_summary['DC_Fast_Coverage_Pct'] = ev_clean.groupby('State').apply(
    lambda x: (x['EV DC Fast Count'] > 0).mean() * 100
).values

# Create network summary
network_summary = ev_clean['EV Network'].value_counts().reset_index()
network_summary.columns = ['Network', 'Station_Count']
network_summary['Percentage'] = (network_summary['Station_Count'] / len(ev_clean) * 100).round(1)

# Save files for Tableau
print(" Saving files for Tableau:")

# Main dataset
tableau_data.to_csv('ev_stations_for_tableau.csv', index=False)
print(f" Main dataset: ev_stations_for_tableau.csv ({len(tableau_data):,} records)")

# State summary
state_summary.to_csv('state_summary_for_tableau.csv', index=False)
print(f" State summary: state_summary_for_tableau.csv ({len(state_summary)} states)")

# Network summary
network_summary.to_csv('network_summary_for_tableau.csv', index=False)
print(f" Network summary: network_summary_for_tableau.csv ({len(network_summary)} networks)")

# Display data previews
print(f"\n Tableau Main Dataset Preview:")
print(f"Shape: {tableau_data.shape}")
print(tableau_data.head(3))

print(f"\n State Summary Preview:")
print(state_summary.head(5))

print(f"\n Network Summary Preview:")
print(network_summary.head(5))

print(f"\n READY FOR TABLEAU!")
print(" Files created:")
print("1. ev_stations_for_tableau.csv - Main dataset for maps and detailed analysis")
print("2. state_summary_for_tableau.csv - State-level aggregations for comparisons")
print("3. network_summary_for_tableau.csv - Network analysis data")
print("\n Next: Import these CSV files into Tableau to create your dashboard!")
```
